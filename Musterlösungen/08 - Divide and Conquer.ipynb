```python
import random


def minMax(ls):
    if len(ls) == 1:
        min = ls[0]
        max = ls[0]
        return min, max
    else:
        if len(ls) == 2:
            if ls[0] < ls[1]:
                min = ls[0]
                max = ls[1]
            else:
                min = ls[1]
                max = ls[0]
            return min, max
        else:
            return minMaxMerge(minMax(ls[:int(len(ls)/2)]), minMax(ls[int(len(ls)/2):]))

        
def minMaxMerge(left, right):
    minLeft, maxLeft = left
    minRight, maxRight = right
    if minLeft < minRight:
        min = minLeft
    else:
        min = minRight
    if maxLeft > maxRight:
        max = maxLeft
    else:
        max = maxRight
    return min, max
```
