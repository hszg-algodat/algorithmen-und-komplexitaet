    ```python
    def reverse(arr):
        for i in range(int(len(arr) / 2)):
            arr[i], arr[len(arr) - 1 - i] = arr[len(arr) - 1 - i], arr[i]
        return arr
    ```

3. Eine Folge von $n$ ganzen Zahlen $p(1), p(2), \dotsc , p(n)$ ist gegeben. Jedes Element der Folge ist verschieden und erfüllt $1 \leq p(x) \leq n$.
    
    a) 
    Schreiben Sie einen Algorithmus, der für jedes $x$ mit $1 \leq x \leq n$ eine ganze Zahl $y$ findet, so dass $p(p(y)) = x$.

    c)
    Falls der Algorithmus nicht in $\mathcal{O}(n)$ läuft, finden Sie einen Algorithmus, der dies in $\mathcal{O}(n)$ tut.
